```python
import pandas as pd
```

```python
path = "/Users/rx32940/Desktop/pbio8350TermProject/"
file = path + "iqtree/interrogans_host_unrooted_2.treefile"

# read in the tree file
with open(file) as f: 
    tree_file=f.read()


# host metadata to dict
host_metadata = pd.read_csv(path + "host_only_interrogan.csv", header = None).iloc[:,0:2]
host_metadata.columns = ["Biosample","Host"]
host_metadata
host_dict = host_metadata.set_index("Biosample").to_dict()["Host"]
host_dict
loc_dict = pd.read_csv(path + "loc_only_interrogan.csv", index_col = 0, header = None).to_dict()[1]
loc_dict

samn_loc=pd.DataFrame()
for key in host_dict:
    if key in loc_dict: # if the isolate with host metadat also has loc metadata
        loc = pd.DataFrame({"Biosample":[key],"Location":[loc_dict[key].replace(":","_")]}) # ":" confuses with tree file
        samn_loc = samn_loc.append(loc, ignore_index = True)
        
host_metadata = host_metadata.merge(samn_loc,how="left")

# write csv file for isolates with host(major) and location metadata 
# host_metadata.to_csv(path + "host_loc_interrogan.csv",header = True, index=False)

for key in host_dict:
    samn_host = key + "|" + host_dict[key]
    if key in loc_dict: # if loc metadata is available
        print(loc_dict[key].replace(": ","_").replace(":","_").replace(", ","_"))
        samn_host = samn_host + "|" + loc_dict[key].replace(": ","_").replace(":","_").replace(", ","_")
        tree_file = tree_file.replace(key, samn_host)
    else:
        tree_file = tree_file.replace(key, samn_host)
tree_file
    
```

```python

with open(path + "iqtree/interrogans_host_unrooted_host_loc.treefile","w") as f:
        f.write(tree_file)

```
